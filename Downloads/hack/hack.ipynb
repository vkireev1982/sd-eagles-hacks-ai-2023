```python
import os
import pandas as pd
import json
```

```python
df=pd.DataFrame()
path='data'
fnames=list(filter(lambda x: x.split('.')[1]=='json',os.listdir(path)))
for fn in fnames:
    full_fn=os.path.join(path,fn)
    with open(full_fn,'r') as f:
        j=json.load(f)
    df=pd.concat([df,pd.json_normalize(j['ovens']).\
                assign(day=fn.split('.')[0]).\
                explode(column='operations').\
                explode(column='working_temps')])
df
```
